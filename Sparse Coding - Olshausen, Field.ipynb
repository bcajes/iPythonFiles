- Here we investigate the proposal (Field cites) that *a coding strategy that maximizes sparseness is sufficient to account for these properties*. We show that a learning algorithm that attempts to find sparse linear codes for natural scenes will develop a complete family of localized, oriented, bandpass receptive fields, similar to those found in the primary visual cortex. The resulting sparse image code provides a more efficient representation for later stages of processing because *it possesses a higher degree of statistical independence among its outputs*. 


- Basic assumption: *an image* $I(x,y)$ *can be represented in terms of a linear superposition of (not necessarily orthogonal) basis functions*, $\phi_i(x,y)$:
$$ I(x,y) = \sum_i a_i \phi_i(x,y) $$
- Image code: determines by the choice of the basis functions, $\phi_i$. 
- The coefficients, $a_i$ are dynamic variables that change from one image to the next. 
- The goal of efficient coding: find a set of $\phi_i$ that forms a complete code (that is, spans the image space) and results in the coefficient values being as statistically independent as possible over an ensemble of natural images. 
    - Why statistically independence: explored in 9, 12 and 19
        19. Barlow (1989) *Unsupervised Learning* http://www.mitpressjournals.org/doi/abs/10.1162/neco.1989.1.3.295#.VOZHEVqrkxI
    - *it provides a strategy for extracting the intrinsic structure in sensory signals*

- via PRINCIPAL COMPONENT ANALYIS:
    1. Sanger (1989)
    2. Hancock et al (1992)
    20. Linsker (1998) 
    - Goal: to find a set of mutually orthogonal basis functions that capture the directions of maximum variance in the data and for which the cooefficients are pairwise decorrelated, $\langle a_i a_j \rangle = \langle a_i \rangle \langle a_j \rangle$
    - Receptive fields made from PCA are: 
        - not localized
        - majority do not resemble any known cortical receptive fields
    - In general, PCA is appropriate when:
        - capturing the structure of data that are well described by a gaussian cloud
        - OR in which the linear pairwise correlations are the most important form of statistical dependence in the data.
    - Natural scenes:
        - contain many higher-order forms of statistical structure
        - good reason to velieve they form an extremely non-gaussian distribution that is not at all well captured by orthogonal components - Field (1994)
        - Lines, edges, esp curved and fractal-like edges cannot be characterized by linear pairwise statistics
            21. Olshausen and Field (1996) (this paper)

- "The existence of any statistical dependencies among a set of variables may be discerned whenever the joint entropy is less than the summ of individual entropies, $H(a_1, a_2, \dots, a_n) < \sum_i H(a_i)$, otherwise the two quantities will equal."
- Assumption: we have som way of ensuring that information in the image (joint entropy) is preserved
    - then, a possible strategy for reducing statistical dependencies is to lower the individual entropies, $H(a_i)$, as much as possible.
- In terms of Barlow $\implies$ we seek a *minimum-entropy code*. 

#### Main conjecture

- natural images have "sparse structure" - **a given image can be represented in terms of a small number of descriptors out of a large set.**
    - "... so we shall seek a specific form of low-entropy code in which **the probability distribution of each coefficient's activity is unimodal and peaked around zero**."
    
- ** The search for a sparse code**
    - an optimization problem with the following cost function:
        $$ E = - [\text{preserve information}] - \lambda[\text{sparseness of } a_i]$$
        where $\lambda$ is a positive constant that determines the importance of the second term relative to the first.
        - **1st term**: measures how well the code describes the image, which we choose this to be the mean square of the error between the actual image and the reconstructed image:
        $$ [\text{preserve information}] = -\sum_{x,y} \bigg[ \underbrace{I(x,y)}_\text{actual image} - \underbrace{\sum_i a_i \phi_i(x,y)}_\text{reconstructed image} \bigg]^2$$
        - **2nd term**: assesses the sparseness of the code for a given image by assigning a cost depending on how activity is distributed among the coefficients: those representations in which activity is spread over many coefficients **should occur at a higher cost** than those in which only a few coefficients carry the load. 
            - to meet this criterion, the cost function takes the sum of each coefficient's activity passed through a nonlinear function $S(x)$:
            $$ [\text{sparseness of } a_i] = - \sum_i S \bigg(\frac{a_i}{\sigma} \bigg)$$
            - $\sigma$ is a scaling constant
            - $S(x)$ has been chosen as $ -e^{-x^2}$, $log (1 + x^2)$ and $\lvert \; x \; \rvert$, with all yielding qualitatively similar results
                - ** reasoning of these $S(x)$ function choices:** they will favor among activity states with equal variance those with the fewest number of non-zero coefficients.

- **Learning:** accomplished by minimizing the total cost functional, $E$. 
    - For each image representation, $E$ is minimized with respect to the $a_i$ (coefficients). 
    - The $\phi_i$ (basis/dictionary functions) then evolve by gradient descent (http://en.wikipedia.org/wiki/Gradient_descent) on $E$ averaged over many image presentations.

   1. for a given image, the coefficients $a_i$ are determined from the equilibrium solution to the differential equation:
   $$ \dot{a_i} = b_i - \sum_j C_{ij} a_j - \frac{\lambda}{\sigma} S' \frac{a_i}{\sigma} $$
   2. where:
   $$ b_i = \sum_{x,y} \phi_i(x,y)I(x,y) \;\;\; \text{and} \;\;\; C_{ij} = \sum_{x,y} \phi_i(x,y)\phi_j(x,y)$$
   3. giving the learning rule for updating the $\phi$
   $$ \Delta \phi_i(x_m,y_n) = \eta \bigg < a_i \bigg[ I(x_m,y_n) - \hat{I}(x_m,y_n)  \bigg]    \bigg >$$
   4. where $\hat{I}$ is the reconstructed image, $\hat{I}(x_m,y_n) = \sum_i a_i \phi_i(x_m,y_n)$, and $\eta$ is the **learning rate**. 
   5. ** the dynamics of $a_i$ and the learning rule of the $\phi_i$ have a local network implementation.**
   6. *** The algorithm is seeking a set of $\phi_i$ for which the $a_i$ can tolerate "sparsification" with minimum reconstruction error.***
       - * the algorithm allows for the basis functions to be **overcomplete** - more basis functions than meaningful dimensions in the input - and non-orthogonal (Daugman 1990)*, without reducing the degree of sparseness in the representation. 
       - Because, the sparseness cost function, $S$, forces the system to choose, in the case of overlaps, which basis functions are most effective for describing a given structure of the image.
