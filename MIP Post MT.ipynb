- **Let's formalize this a bit** (See notes for figure/coordinate system)
    - done by [RADON TRANSFORM](http://en.wikipedia.org/wiki/Radon_transform)
    $$ \vec{\omega} = \begin{pmatrix} t cos \theta \\ t sin \theta \end{pmatrix} + s \begin{pmatrix} - sin \theta \\ cos \theta \end{pmatrix}$$  
    $$ R \bigg( f(x,y) \bigg) = \int_{- \infty} ^\infty f \bigg( \underbrace{\big(t cos \theta - s sin \theta \big)}_\text{x}, \underbrace{\big(t sin \theta + s cos \theta \big)}_\text{y} \bigg) \mathrm{d} s $$
