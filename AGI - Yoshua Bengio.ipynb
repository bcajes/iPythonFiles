#### Non-distributed representations

- Prototypical ML training, how it has been done for decades
- All these measures do is they take the input space and break it up into a number of regions
    - this is very powerful, they can break up the input space into as many regions as you need, as long as you have enough examples to tell where these regions are and what the output should be for each of these regions
- Clustering, Nearest-Neighbors (need *one* example per region), RBF (radial basis function support vector machiens) SVMs, local non-parametric density estimation & prediction, decision trees, etc.
- Parameters for each distinguishable regions
- Q: How many regions can you get for how many examples?
    - the # of distinguishable regions is linear in # of parameters
- No non-trivial generalization to regions without examples

#### The need for distributed representations

- Another way of breaking up input space: Factor models, PCA (principal component analysis), RBMs, Neural networks, sparse coding, deep learning, etc.
- Each parameter influences many regions, not just local neighbors
- the # of distinguishable regions grows almost exponentially with # of parameters
- GENERALIZE NON-LOCALLY TO NEVER-SEEN REGIONS (in the prior example, you would REQUIRE an example of a region in order to establish classifier there. In THIS CASE, I do not need to see an example of a combination in order to be able to classify a novel example of that set correctly.)
- Therefore, while not independent in a statistical sense, they are independent in the case that I do not need to learn ALL FEATURES in order to construct a successful classifier
- Non-mutually exclusive features/attributes create a combinatorially large set of distinguishable features
- Example: 2D input and need to learn 3 binary features (imagine they are just linear classifiers)

#### Issues with Back-Propagation

- Trying to get rid of Back-Prop
- relies on infinitesimal changes in order to make changes and learn
- Over very deep nets or recurrent nets with many steps, non-linearities compose and yield sharp non-linearity -> gradient vanish or explode (cant use gradient descent here)
- Trainging deeper nets: harder optimization
- In the extreme of non-linearity: discrete functions, can't use back-propagation
- Not biologically plausible


#### How Brains Might Learn Without Backpropagation

- Two principles:
    - The past tries to match the future: prediction (neurons are trying to predict their future values)
    - The future tries to match the past: reconstruction (idea, neurons are also trying to match their past)
    - *It is not clear if these should be on same or different units*
- Plus: observations being clamped (not always)
- Does not depend on the form of activation function, tied symmetric weights, differentiability of anything, using rates vs. spikes, etc.
