### Why are magnets magnetic?

- All matter has some king of magnetic property because electrons ($e^{-}$) have "spin"

- A classical way of thinking about spin = angular momentum (has a rotation and a preferred direction)

- within a given volume, the magnetization $\vec{\mathbf{M}}$ is the vector sum of all spins

- for most materials that are NOT in a magnetic field, spins are in all directions $\to$ NON-MAGNETIZED (magentization sum = 0).

### Magnetic properties of materials

- Whay are these spin configurations random in the first place? Where does this randomization come from? 
    - randomization occurs due to *thermal fluctuations*, which are greater than the internal interactions
    - how many align depend on the strength of the magnetic field (the stronger $\vec{\mathbf{B}}$ - more likely to align - can become MAGNETIZED, since th vector sum will be finite and non-zero)
    - http://en.wikipedia.org/wiki/Thermal_fluctuations

$$\vec{\mathbf{M}} = \chi \vec{\mathbf{B}}$$
- where $\chi$ is the *magnetic susceptibility* (a dimensionless proportionality constant that indicates the degree of magnetization of a material in response to an applied magnetic field)(http://en.wikipedia.org/wiki/Magnetic_susceptibility)
- If $ \chi > 0 $, the material is paramagnetic (http://en.wikipedia.org/wiki/Paramagnetism)
    - $\vec{\mathbf{M}}$  is *parallel* to $\vec{\mathbf{B}}$
    - attracts magnetic field lines
    - increases the field inside
    - Paramagnetic materials: Aluminum, OXYGEN, DEOXYHEMOGLOBIN
- If $ \chi < 0 $, the material is diamagnetic (http://en.wikipedia.org/wiki/Diamagnetism)
    - $\vec{\mathbf{M}}$  is *opposite* to $\vec{\mathbf{B}}$
    - repells magnetic field lines
    - weakens the field inside
    - Diamagnetic materials: Mercury, copper, carbon, HYDROGEN, OXYHEMOGLOBIN
    
- fMRI is based on measuring the difference between:
oxyhemoglobin (DIAMAGNETIC) $\parallel$ to $\vec{\mathbf{B}}$


 $$oxyhemoglobin \;\;(DIAMAGNETIC) =
  \begin{cases}
   deoxyhemoglobin \;\;(PARAMAGNETIC) \\
   oxygen \;\; (PARAMAGNETIC)
  \end{cases}$$
  
  
- For both paramagnetic and diamagnetic materials, $\vec{\mathbf{M}} = 0$ when $\vec{\mathbf{B}} = 0$(no magnetic hysteresis in these cases).
- In the case of ferromagnetic materials, (http://en.wikipedia.org/wiki/Ferromagnetism) $$\vec{\mathbf{M}} \neq \chi \vec{\mathbf{B}}$$
    - *FERROMAGNETIC HYSTERESIS*
    - *Critical temperature (Curie temperature)* - where the magnetization disappears
- Perfect diamagnet
    - No magnetic fields inside
    - Superconductors (Type I) (http://en.wikipedia.org/wiki/Superconductivity) (http://en.wikipedia.org/wiki/Superconducting_magnet)
    - Meissner-Odesenfeld effect - the expulsion of a magnetic field from a superconductor during its transition to the superconducting state. (http://en.wikipedia.org/wiki/Meissner_effect)
    - If you put a superconductor into a large enough magnetic field - it is no longer superconductive
    - fMRI - superconductor creates its own magnetic field, which can not be increased as much as you want, or else the superconductive $\vec{\mathbf{B}}$ "will kill itself".

### Gyroscopes in a Force Field

- without spin or rotation -> FALLS DOWN
- with spin/rotation -> path around central axis -> PRECESSION
- Precession occurs at certain frequency -> LAMOR FREQUENCY - (http://en.wikipedia.org/wiki/Larmor_precession)
    - In physics, Larmor precession (named after Joseph Larmor) is the precession of the magnetic moments of electrons, muons, and all leptons with magnetic moments, which are quantum effects of particle spin, atomic nuclei, and atoms about an external magnetic field. The magnetic field exerts a torque on the magnetic moment,

$$\vec{\Gamma} = 
\vec{\mu}\times\vec{B}=
\gamma\vec{J}\times\vec{B}$$

where $\vec{\Gamma}$ is the torque, $\vec{\mu}$ is the magnetic dipole moment, $\vec{J}$ is the angular momentum vector, $\vec{B}$ is the external magnetic field, $\times$ symbolizes the cross product, and $\ \gamma$ is the gyromagnetic ratio which gives the proportionality constant between the magnetic moment and the angular momentum.

- in imaging, we are dealing with $e^{-}$ with spin.

### Magnetization $\vec{\mathbf{M}}$ in a magnetic field $\vec{\mathbf{B}}$

- When $\uparrow \vec{\mathbf{M}}$  $\uparrow \vec{\mathbf{B}}$ PARALLEL $\to$ LOW ENERGY (Paramagnetic)
- When $\downarrow \vec{\mathbf{M}}$  $\uparrow \vec{\mathbf{B}}$ OPPOSITE/ANTI-PARALLEL $\to$ HIGH ENERGY (Diamagnetic)

- Where the energy difference ($\Delta E$) is proportional to $|\vec{\mathbf{B}}|$

## Quantitative Dynamics of Magnetization (Bloch equations)

Total system (system of 1st order linear differential equations: 

$$ \begin{pmatrix} 
  \dot{m_X} \\ \dot{m_Y}\\ \dot{m_Z}
 \end{pmatrix} = -\frac{g \mu}{\hbar} \vec{\mathbf{m}} \times \vec{\mathbf{B}} +  \begin{pmatrix} 
  -\frac{1}{T_2}{m_Y} \\ -\frac{1}{T_2}{m_X}\\ \frac{M_0 - m_Z}{T_1}
 \end{pmatrix}$$
 
- where:
    - g: g-factor (a dimensionless quantity which characterizes the magnetic moment and gyromagnetic ratio of a particle or nucleus. http://en.wikipedia.org/wiki/G-factor_(physics) )
    - $\hbar = \frac{h}{2 \pi}$, where h is Planck's constant (http://en.wikipedia.org/wiki/Planck_constant)
    - $\mu$: Bohr's magneton (describes the magnetic properties of the proton) (http://en.wikipedia.org/wiki/Bohr_magneton)

- we can begin our analysis of these equations under an assumption that g = 1

- $m_X,m_Y,m_Z \implies$ components of the magnetization
- $M_0 \implies$ magnetization at equilibrium

- First, let's deal with the cross-product, http://en.wikipedia.org/wiki/Cross_product

$$ \vec{\mathbf{m}} \times \vec{\mathbf{B}} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ m_X & m_Y & m_Z \\ B_X & B_Y & B_Z \end{vmatrix} \begin{matrix} \hat{i} & \hat{j} \\ m_X & m_Y \\ B_X & B_Y \end{matrix} = \begin{pmatrix} m_Y B_Z - m_Z B_Y \\ m_Z B_X - m_X B_Z \\ m_X B_Y - m_Y B_X \end{pmatrix} $$

- Giving, when we assume that g = 1, we can separate the three equations: 

$$ \dot{m_X} = - m_Y B_Z + m_Z B_Y - \frac{1}{T_2} m_X $$
$$ $$
$$ \dot{m_Y} = - m_Z B_X + m_X B_Z  - \frac{1}{T_2} m_Y$$
$$ $$
$$ \dot{m_Z} = - m_X B_Y + m_Y B_X - \frac{M_0 - m_Z}{T_1}$$

- We define $B_Z$ as constant (we have a driven system if we keep the z-component constant) 
- We apply a RF-pulse along one axis $\perp$ to $B_Z$: $ B_X = Acos(\gamma t)$, where $\gamma$ is the Lamor frequency.
    - by choosing the x-direction as the one for our RF-pulse, we can now remove $B_Y$ from the zero by $B_Y = 0$.

- Which shows us that we have a *non-autonomous* system of equations (time appears explicitly) $\implies$ THIS SUCKS http://en.wikipedia.org/wiki/Non-autonomous_system_(mathematics)
    - If you can't solve a system, simplify it

- If we set $B_X$ = 0, we have no RF pulse:

$$ 2D \;\; System \;\;= 
  \begin{cases}
    \dot{m_X} = - m_Y B_Z  - \frac{1}{T_2} m_X  \\
   \dot{m_Y} =  + m_X B_Z  - \frac{1}{T_2} m_Y
  \end{cases}$$

$$ 1D \;\; System \;\;= 
  \begin{cases}
    \dot{m_Z} = \frac{M_0}{T_1} - \frac{m_Z}{T_1} 
  \end{cases}$$

### The z-component

- DECOUPLED: $m_Z$ does not depend on $m_X$ and $m_Y$, and vice versa. 

$$\dot{m_Z} = \frac{M_0}{T_1} - \frac{m_Z}{T_1}$$
$$ \frac{\mathrm{d}m_Z}{\mathrm{d}t} = \frac{1}{T_1} ( M_0 - m_Z)$$
$$ \int \frac{1}{M_0 - m_Z} \mathrm{d}m_Z = \int \frac{1}{T_1} \mathrm{d}t $$
$$ -ln(M_0 - m_Z) + C = \frac{t}{T_1} + C $$
$$ e^{-ln(C)} = e^{-\frac{t}{T_1} + C} $$
$$ M_0 - m_Z = e^{-\frac{t}{T_1}} e*{C} = C e^{-\frac{t}{T_1}}$$
$$ m_Z = M_0 + C e^{-\frac{t}{T_1}}$$

- This is the *general solution* of $\dot{m_Z} = \frac{M_0}{T_1} - \frac{m_Z}{T_1} $
- It has an open parameter $C$, which is determined by initial conditions
    - at $t = 0$, $\vec{m}$ has an angle $\alpha$ with respect to the z-axis

$$ m_Z(t=0) = M_0 cos \alpha = M_0 + C$$
$$ \implies C = M_0 (cos \alpha - 1)$$
$$ m_Z(t) = M_0 + M_0(cos \alpha - 1) e^{-\frac{t}/{T_1}} $$ 
$$ $$
$$ \boxed{m_Z(t) = M_0 \left\{1 + (cos \alpha - 1) e^{-\frac{t}{T_1}}\right\}}$$

- and in the special case of $\alpha = 90^{\circ}$, as in the rotation into the xy-plane:

$$ \boxed{m_Z(t) = M_0 \left\{1 + e^{-\frac{t}{T_1}}\right\}}$$

- which is exactly the longitudinal spin axis relaxation, $T_1$.
- Note that for the case of $\alpha = 0^{\circ}$, or when the magnetization vector has relaxed back into its equilibrium position, $ m_Z = M_0 $.

### The xy-components

$$ \begin{matrix} \dot{m_X} = - m_Y B_Z  - \frac{1}{T_2} m_X \\ 
 \dot{m_Y} =  + m_X B_Z  - \frac{1}{T_2} m_Y \end{matrix} \implies \begin{pmatrix} \dot{m_X} \\ \dot{m_Y} \end{pmatrix} = 
   \overbrace{\begin{pmatrix} -\frac{1}{T_2} & -B_Z \\ B_Z & -\frac{1}{T_2} \end{pmatrix}}^\text{$\mathrm{A}$} \begin{pmatrix} {m_X} \\ {m_Y} \end{pmatrix}$$
   
- We know from Nonlinear Dynamics that the *General Solution* for a 2D system of this kind is:

$$ \begin{pmatrix} {m_X(t)} \\ {m_Y(t)} \end{pmatrix} = c_1 e^{\lambda_1 t} \mathbf{v}^{(1)} + c_2 e^{\lambda_2 t}\mathbf{v}^{(2)} $$

- where $\lambda_{1,2}$ and $\mathbf{v}^{(1),(2)}$ are the eigenvalues and eigenvectors of $\mathrm{A}$.  http://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors
    - take the determinant and set to zero: $ \big | \mathrm{A} - \lambda \mathrm{I} \big |  = 0$ 
    - setting the determinant to zero makes these two equations linearly dependent, we can find the $\lambda_{1,2}$ such that we are not left with only the trivial solution, $ 0 = 0 $. 

$$ \big | \mathrm{A} - \lambda \mathrm{I} \big | = \begin{vmatrix} -\frac{1}{T_2} - \lambda & -B_Z \\ B_Z & -\frac{1}{T_2} - \lambda \end{vmatrix} = \overbrace{ \lambda^2 + \frac{2}{T_2} \lambda + \frac{2}{T_2^2} + B_Z^2 }^\text{characteristic polynomial} = 0$$

$$ \lambda_{1,2} = \frac{1}{2} \left\{ -\frac{2}{T_2} \pm \sqrt{\frac{4}{T_2^2} - \frac{4}{T_2^2} - 4 B_Z^2 } \right\} $$

$$ \boxed{\lambda_{1,2} = -\frac{1}{T_2} \pm i B_Z} $$


### Eigenvectors

- In general: $\mathrm{A} \vec{v} = \lambda \vec{v}$
    - multiplying a matrix $\mathrm{A}$ will not change its direction; $\mathrm{A}$ performs only a scaling but no rotation!

- The system of 2 linear equations:
    
    $$ \begin{pmatrix} -\frac{1}{T_2} & -B_Z \\ B_Z & -\frac{1}{T_2} \end{pmatrix} \begin{pmatrix} v_X \\ v_Y \end{pmatrix} = 
    \overbrace{ \bigg( -\frac{1}{T_2} \pm i B_Z \bigg)}^\text{$\lambda_{1,2}$} \begin{pmatrix} v_X \\ v_Y \end{pmatrix} $$
    
    $$ -\frac{1}{T_2} v_X - B_Z v_Y = \lambda_{1,2} v_X $$
    $$ B_Z v_X - \frac{1}{T_2} v_Y = \lambda_{1,2} v_Y $$
    $$ $$ 
    $$ \begin{matrix} homogeneous \;\; system \\ of \;\; linear \;\; equations \end{matrix} = \begin{cases}
    \bigg(-\frac{1}{T_2} - \lambda_{1,2} \bigg) v_X - B_Z v_Y = 0  \\
   B_Z v_X - \bigg(\frac{1}{T_2} + \lambda_{1,2} \bigg) v_Y = 0
  \end{cases} $$
  
- a homogeneous system always has the trivial solution
- substituting in $\lambda_{1,2} = -\frac{1}{T_2} \pm i B_Z$:

$$ \bigg( -\frac{1}{T_2} - \lambda_{1,2} \bigg) v_X - B_Z v_Y = 0 \implies \bigg( -\frac{1}{T_2} + \frac{1}{T_2} \mp i B_Z \bigg) v_X - B_Z v_Y = 0 $$ 
$$ B_Z v_X - \bigg(\frac{1}{T_2} + \lambda_{1,2} \bigg) v_Y = 0 \implies B_Z v_X - \bigg(\frac{1}{T_2} -\frac{1}{T_2} \mp i B_Z \bigg) = 0 $$
$$ $$
$$ \bigg(\mp i B_Z \bigg) v_X - B_Z v_Y = 0 \implies \mp \; i B_Z  v_X =  B_Z v_Y \implies \mp \; i  v_X =  v_Y$$
$$ B_Z v_X - \bigg(\mp i B_Z \bigg)v_Y = 0 \implies \mp \; i B_Z  v_Y =  B_Z v_X \implies \mp \; i  v_Y =  v_X$$ 
$$ $$
$$ \implies \mathbf{v}^{(1),(2)} = \begin{pmatrix} 1 \\ \mp \; i \end{pmatrix} $$

### General Solution

$$ \begin{pmatrix} {m_X(t)} \\ {m_Y(t)} \end{pmatrix} 
= c_1 e^{(-\frac{1}{T_2} + i B_Z) t} \begin{pmatrix} 1 \\ - i \end{pmatrix} 
+ c_2 e^{(-\frac{1}{T_2} -i B_Z) t} \begin{pmatrix} 1 \\  i \end{pmatrix} $$

- we know that the output of the general solution is real, however the function is still in terms of the imaginary $i$. Therefore, we must compensate by making the coefficients $c_1$ and $c_2$ complex conjugates $c$ and $c^{*}$

$$ \begin{pmatrix} {m_X(t)} \\ {m_Y(t)} \end{pmatrix} 
= \underbrace{c}_\text{$c_r + i c_i$} e^{(-\frac{1}{T_2} + i B_Z) t} \begin{pmatrix} 1 \\ - i \end{pmatrix} 
+ \underbrace{c^{*}}_\text{$c_r - i c_i$} e^{(-\frac{1}{T_2} -i B_Z) t} \begin{pmatrix} 1 \\  i \end{pmatrix} $$

- Giving,

$$ m_X(t) = c e^{(-\frac{1}{T_2} + i B_Z) t} + c^{*} e^{(-\frac{1}{T_2} -i B_Z) t} = e^{-\frac{t}{T_2}} \bigg ( c e^{i B_Z t} + c^{*} e ^{-i B_Z t} \bigg )  $$
$$ m_Y(t) = - i c e^{(-\frac{1}{T_2} + i B_Z) t} + i c^{*} e^{(-\frac{1}{T_2} -i B_Z) t} = i e^{-\frac{t}{T_2}} \bigg ( c^{*} e ^{-i B_Z t} - c e^{i B_Z t} \bigg )$$ 

- First, the x-component:

$$ m_X(t) = e^{-\frac{t}{T_2}} \bigg ( (c_r + i c_i) e^{i B_Z t} + (c_r - i c_i) e ^{-i B_Z t} \bigg )  $$

- Note the identity: $ e^{i B_Z t} = cos B_Z t + i sin B_Z t $

$$ m_X(t) = e^{-\frac{t}{T_2}} \bigg ( (c_r + i c_i) (cos B_Z t + i sin B_Z t) + (c_r - i c_i) (cos B_Z t - i sin B_Z t) \bigg ) $$ 

$$ \implies m_X(t) = e^{-\frac{t}{T_2}} \bigg ( 2 c_r cos B_Z t - 2 c_i sin B_Z t  \bigg ) $$


- Then, the y-component:

$$ m_Y(t) = i e^{-\frac{t}{T_2}} \bigg ( (c_r - i c_i) e ^{-i B_Z t} - (c_r + i c_i) e^{i B_Z t} \bigg )$$ 

$$ m_Y(t) = i e^{-\frac{t}{T_2}} \bigg ( (c_r - i c_i) (cos B_Z t - i sin B_Z t) - (c_r + i c_i) (cos B_Z t + i sin B_Z t) \bigg )$$ 

$$ \implies m_Y(t) = e^{-\frac{t}{T_2}} \bigg ( 2 c_r sin B_Z t + 2 c_i cos B_Z t  \bigg ) $$

- Then, to solve for the constants $c_r$ and $c_i$, we have the initial condition $m_X(t=0) = 0$ and $m_Y(t=0) = M_0$

$$ m_X (t = 0) = 2 c_r = 0 \implies c_r = 0 $$

$$ m_Y(t = 0) = 2 c_i = M_0 \implies c_i = \frac{M_0}{2} $$

- Finally giving,

$$ \boxed{\begin{matrix}
 m_X(t) = - M_0 e^{-\frac{t}{T_2}} sin B_Z t \\
 m_X(t) = M_0 e^{-\frac{t}{T_2}} cos B_Z t \end{matrix} }$$

- precession in the xy-plane, which decays with time constant $T_2$.
- the presence of an exponential in front of a periodic function $\implies$ DAMPED OSCILLATION http://hyperphysics.phy-astr.gsu.edu/hbase/oscda.html

- NOTE: the equation above would imply that the frequency $ \omega = B_Z$, but remember that we set the original system's coefficient $\frac{g \mu}{\hbar} = 1$. So really the actual precession frequency: 

$$ 2 \pi \nu = \omega = \frac{g \mu}{\hbar} B_Z $$
