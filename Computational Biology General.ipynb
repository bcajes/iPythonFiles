            - Neuronal (Model and find replicators in the brain)

    - *What is natural selection?*
        - John Maynard Smith - units of evolution multiply - > *but this is not sufficient to describe evolultion, since there is no selection. Fire multiplies but is not selected upon*
            - you also need heredity (offspring resemble parents more than others) 
            - you also need *variation* (differences within a generation - variability is differnces over multiple generations)
            - **Therefore**, units of evolution have multiplication, variation and heredity, with selection yielding higher fitness
        - *We need to distinguish natural selection from many other algorithms that **are not** natural selection*

      - *A Darwinian Cognitive Architecture* 
          - aim: black box with different sensory input able to plug into it
          - schism between two camps: physical symbol systems (Fogel) and connectionist/neural network people
              - Q: what is the minimum structure needed for general intelligence? 
              - Fogel: you need a physical symbol system - connectionist systems do not have atomic systems with rules of linkages that can be recombined
          - Consists of functional 'atoms' capable of replication (a neural microcircuit - piece of cortex)
              - atom could be a single classification hierarchy structure from ML
              - atom = a neuronal group in cortex with certain functional properties (response characteristics, processing abilities)
              - replication of atoms is evident during cortical map plasticity where cortical representations grow and shrink
          - These link to form 'molecules' capable of replication
              - molecules are functional circuits. They are implemented by cortex (atom) - Basal Ganglia (register/router/edge) - cortex (atom) bindings
              - BG could coordinated atoms switching on and off other atoms
          - 'Molecules' aggregate into function 'compartments' capable of replication
              - these are co-activated groups of molecules that control behavior as a unit. They are an important unit of selection in the brain
              - these 'compartments' could be probabilistic, but they are the 'current' modules to which reward is attributed

      - **Evolving Games/Fitness Functions**
          - The algorithm didn't invent the game to play. We imposed it.
          - Can an algorithm decide which games to play?
          - How should it decide what a good game to play is?
          - As academics we also invent our research problems and try to solve them
          - Clever children like clever academics choose the right games to play
          - Q: How do you choose the games to play?
          - What is a good game/goal for a baby?
              - Not too hard
                  - all solutions have low fitness on the game (floor effect)
              - Not to easy
                  - all solutions have high fitness on the game (ceiling effect)
              - just right **(progress is being made, and further progress is predicted)**
                  - maximum variance in the quality of modified (best guessed) novel solutions
                  - This is deeply related to **Fisher's fundamental theorem of natural selection: The rate of fitness increase is proportional to the fitness variance** (the more differnce there is in the fitnes of a population the greater rate of fitness change)
                  - we can evolve games with this variance as a metric
                  - but it has some stipulations that dont actually occur in real systems (since most evolutionary change is neutral)
                  - Given some assumptions, e.g. no linkage disequilibrium, no frequency-dependent selection, no epistasis (yeah right)
                  - neutral exploration is more common, resulting in **punk eak**, and making evolutionary progress hard to predict.
       - **General Minimal Criteria for Good Games**
           - Absolute criteria:
               - A game is not worth playing unless progress is being made on it, or if progress is not predicts [Progress criterion, Oudayer]
               - A game is not worth playing if it likely to kill you, cause you irreparable damage [Variability criterion, Ashby]
           - Relative criteria:
               - a good game allows learning of information useful for real challenges later [Latent learning criterion, Tolman]
               - A good game builds knowledge about the world: allows facts to be learned [Epistemic criterion, Barto]
               - A good game builds a deep understanding about the world, allows generative models (theories) to be learned [Compression criterion, Schmidhuber - interest is the first derivative of compression]
               - A good game allows one to predict more about the future i.e. maximizes that first derivative of predictive information [Predictive information criterion, Der, Mathius] (how much uncertainty about the future can you minimize given the current state + you know the past)
               - A good game allows many novel solutions to be produced quickly [Novelty criterion, Lehman and Stanley] (removed the fitness function, only reward an individual if they produce distant behaviors from other solutions according to some measure) 
               - ** Current problem: open-ended evolution for continual novelty**
